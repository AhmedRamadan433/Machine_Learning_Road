```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Plot Data

```python
np.random.seed(0)
x = np.linspace(0, 100, 100)
y = 5 * x - 12
ry = np.random.normal(scale=80,size=y.shape[0]) + y
# Initilize a DataFrame for the data
data = pd.DataFrame({'x': x, 'y':ry}).sample(frac=1).reset_index()[['index', 'y']].rename({'index':'x'}, axis=1)
plt.plot(x, ry, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(x, ry, 'o')
plt.plot(x, y, 'r-')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Equation of linear legression is :

𝐘 = 𝓦1*𝑋1+𝓦0

Imprt Model from sklearn

```python
from sklearn.linear_model import LinearRegression
model =LinearRegression()
```

Fit data in model

```python
x = data['x'].values.reshape((-1, 1))
y = data['y'].values.reshape((-1, 1))
model.fit(x,y)
```

Get weights of equation

```python
w1=model.coef_
w0=model.intercept_
print(w1,w0)
```

draw predicted line and compare with true line

```python
pred_line = w1[0,0] * x.reshape(-1) + w0[0]
plt.plot(x.reshape(-1), data['y'], 'o')
plt.plot(x.reshape(-1), pred_line, 'r-')
plt.plot(x.reshape(-1), 5 * x - 12, 'b-')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Mean square Erorr metrix

```python
from sklearn.metrics import mean_squared_error
ypred=model.predict(x)
print("MSE : ",mean_squared_error(y,ypred))
```
