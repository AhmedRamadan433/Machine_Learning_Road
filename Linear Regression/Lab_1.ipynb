 **Import Libiraries**
  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Import Data**


```python
data =pd.read_csv('/content/Salary-Data.csv')
data.head()
```

**Plot Data**

```python
import matplotlib.pyplot as plt

plt.plot(data['Experience Years'], data['Salary'], 'o')

plt.title("Relationship Between Experience and Salary")
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

##### prepare Data

```python
y = data['Salary']
y
```

```python
X = data[['Experience Years']]
X
```

##### Modeling Part


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Select Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

Fit Data

```python
model.fit(X_train,y_train)
```

```python
model.coef_ ,model.intercept_
```

```python
y_pred = model.predict(X_test)
y_pred
```

Chek Accuracy

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

**Plot Acutal Data and predicted**

```python
# make array using model coefficients
m = model.coef_[0]
b = model.intercept_

x_values = data['Experience Years'].values
y_manual = m * x_values + b
```

```python
plt.scatter(data['Experience Years'], data['Salary'], color='blue', label='Actual Data')
plt.plot(x_values, y_manual, color='green', linewidth=2, label='model Line: y=mx+b')
plt.title("Linear Regression line")
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.legend()
plt.grid(True)
plt.show()
```
